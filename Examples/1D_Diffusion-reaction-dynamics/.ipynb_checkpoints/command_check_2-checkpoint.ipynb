```python
import numpy as np
import matplotlib.pyplot as plt

# Example data (replace with your actual data)
# Model A
loss_model_a = [0.5, 0.4, 0.32, 0.29, 0.27]
runtime_per_iter_a = 0.8  # seconds
iterations_a = np.arange(1, len(loss_model_a) + 1)
runtime_a = iterations_a * runtime_per_iter_a

# Model B
loss_model_b = [0.6, 0.45, 0.38, 0.35, 0.33]
runtime_per_iter_b = 0.5  # seconds
iterations_b = np.arange(1, len(loss_model_b) + 1)
runtime_b = iterations_b * runtime_per_iter_b

# Plotting
plt.figure(figsize=(8, 5))
plt.plot(runtime_a, loss_model_a, marker='o', label='Model A')
plt.plot(runtime_b, loss_model_b, marker='s', label='Model B')
plt.xlabel("Runtime (seconds)")
plt.ylabel("Test Loss")
plt.title("Test Loss vs Runtime")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Example data (replace with your actual data)
# Vanilla model
loss_vanilla = [0.5, 0.4, 0.32, 0.29, 0.27]
runtime_per_iter_vanilla = 0.8  # seconds
iterations_vanilla = np.arange(1, len(loss_vanilla) + 1)
runtime_vanilla = iterations_vanilla * runtime_per_iter_vanilla

# Latent model
loss_latent = [0.6, 0.45, 0.38, 0.35, 0.33]
runtime_per_iter_latent = 0.5  # seconds
iterations_latent = np.arange(1, len(loss_latent) + 1)
runtime_latent = iterations_latent * runtime_per_iter_latent

# Plotting
plt.figure(figsize=(8, 5))
plt.plot(runtime_vanilla, loss_vanilla, marker='o', label='Vanilla')
plt.plot(runtime_latent, loss_latent, marker='s', label='Latent')
plt.xlabel("Runtime (seconds)")
plt.ylabel("Test Loss")
plt.title("Test Loss vs Runtime")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
