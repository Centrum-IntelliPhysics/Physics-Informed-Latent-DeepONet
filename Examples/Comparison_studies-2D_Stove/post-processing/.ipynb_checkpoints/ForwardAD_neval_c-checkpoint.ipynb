```python
import nbformat
import pandas as pd
import numpy as np
from matplotlib import rc
import matplotlib.pyplot as plt
import numpy as np

import sys
sys.path.append("../../..")
from utils.misc_postprocessing import *

import matplotlib
matplotlib.rcParams['text.usetex'] = True
matplotlib.rcParams['text.latex.preamble'] = r'\usepackage{amsmath}'

# rc('font', **{'family': 'serif', 'serif': ['Computer Modern']})
# rc('text', usetex=True)
```

```python
discretization_list = [8, 16, 32, 64, 128, 256, 512, 1024]
# List of 'neval' values
neval_values = [(i, i) for i in discretization_list]
```

```python
# Loop through each neval values

# Initialize an empty list to store the results  
runtime_per_iter_data = []

# Loop through neval_t and neval_x values
for neval_t, neval_x in neval_values:
    # Paths for notebooks
    a_PI_Vanilla_NO_ForwardADeinsum_path = f'../results/ForwardAD_neval_c/a_Vanilla-NO_ForwardADeinsum/output_neval_t={neval_t}_neval_x={neval_x}.ipynb'
    b_PI_Latent_NO_ForwardADeinsum_path = f'../results/ForwardAD_neval_c/b_Latent-NO_ForwardADeinsum/output_neval_t={neval_t}_neval_x={neval_x}.ipynb'

    # Extract runtime and memory for all notebooks
    a_PI_Vanilla_NO_ForwardADeinsum_runtime, a_PI_Vanilla_NO_ForwardADeinsum_memory = extract_runtime_and_memory_from_notebook(a_PI_Vanilla_NO_ForwardADeinsum_path)
    b_PI_Latent_NO_ForwardADeinsum_runtime, b_PI_Latent_NO_ForwardADeinsum_memory = extract_runtime_and_memory_from_notebook(b_PI_Latent_NO_ForwardADeinsum_path)

    # Append the results to the list
    runtime_per_iter_data.append({
        'neval_t': neval_t,
        'neval_x': neval_x,
        'PI-Vanilla-NO (ForwardAD-einsum) Runtime (sec/iter)': a_PI_Vanilla_NO_ForwardADeinsum_runtime,
        'PI-Vanilla-NO (ForwardAD-einsum) Memory (MB)': a_PI_Vanilla_NO_ForwardADeinsum_memory,
        'PI-Latent-NO (ForwardAD-einsum) Runtime (sec/iter)': b_PI_Latent_NO_ForwardADeinsum_runtime,
        'PI-Latent-NO (ForwardAD-einsum) Memory (MB)': b_PI_Latent_NO_ForwardADeinsum_memory
    })

# Convert the results into a DataFrame for the current latent dimension
df = pd.DataFrame(runtime_per_iter_data)
df.index.name = 'Evaluation'

display(df)
    
```

```python
### RUNTIME (sec/iter)

# Assuming df is the DataFrame we created
n_values = df['neval_t']

# Plotting the four runtime columns
plt.figure(figsize=(6, 4))

# Plot PI-Vanilla-NO
plt.plot(n_values, df['PI-Vanilla-NO (ForwardAD-einsum) Runtime (sec/iter)'], label='PI-Vanilla-NO', 
         marker='o', linestyle='-', color="#1f77b4")
# Plot PI-Latent-NO 
plt.plot(n_values, df['PI-Latent-NO (ForwardAD-einsum) Runtime (sec/iter)'], label='PI-Latent-NO', 
         marker='o', linestyle='-', color="#d62728")

# Setting plot details
plt.xscale('log')  # Logarithmic scale for x-axis
plt.yscale('log') # Logarithmic scale for y-axis
plt.xlabel('Number of collocation points in the solution space ($n_{t}^r \\times n_{\\boldsymbol{x}}^r$)', fontsize=14) # Number of collocation points ($n_c^2$)
# plt.xlabel('Discretization of the solution space ($n_c^2$)', fontsize=14) # Number of collocation points ($n_c^2$)
plt.ylabel('Runtime per iteration (sec/iter)', fontsize=14)
plt.xticks(n_values, [f'{int(n)}$^3$' for n in n_values])  # Custom x-axis labels with superscript
plt.legend(loc='best', fontsize=12)
plt.grid(True)
# plt.title('ForwardAD-einsum')
plt.tick_params(axis='both', labelsize=14)
plt.tight_layout()
# plt.savefig("runtime.pdf") 
plt.show()


### Memory (MB)

# Assuming df is the DataFrame we created
n_values = df['neval_t']

# Plotting the four runtime columns
plt.figure(figsize=(6, 4))

# Plot PI-Vanilla-NO 
plt.plot(n_values, df['PI-Vanilla-NO (ForwardAD-einsum) Memory (MB)'], label='PI-Vanilla-NO', 
         marker='o', linestyle='-', color="#1f77b4")
# Plot PI-Latent-NO  
plt.plot(n_values, df['PI-Latent-NO (ForwardAD-einsum) Memory (MB)'], label='PI-Latent-NO', 
         marker='o', linestyle='-', color="#d62728")

# Setting plot details
plt.xscale('log')  # Logarithmic scale for x-axis
plt.yscale('log') # Logarithmic scale for y-axis
plt.xlabel('Number of collocation points in the solution space ($n_{t}^r \\times n_{\\boldsymbol{x}}^r$)', fontsize=14) # Number of collocation points ($n_c^2$)
# plt.xlabel('Discretization of the solution space ($n_c^2$)', fontsize=14)# Number of collocation points ($n_c^2$)
plt.ylabel('Memory (MB)', fontsize=14)
plt.xticks(n_values, [f'{int(n)}$^3$' for n in n_values])  # Custom x-axis labels with superscript
plt.legend(loc='best', fontsize=12)
plt.grid(True)
# plt.title('ForwardAD-einsum')
plt.tick_params(axis='both', labelsize=14)
plt.tight_layout()
# plt.savefig("memory.pdf") 
plt.show()
```
