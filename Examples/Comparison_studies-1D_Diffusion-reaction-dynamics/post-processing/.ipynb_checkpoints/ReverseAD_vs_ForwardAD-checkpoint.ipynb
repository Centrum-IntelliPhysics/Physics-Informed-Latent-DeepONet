```python
import nbformat
import pandas as pd
import numpy as np
from matplotlib import rc
import matplotlib.pyplot as plt
import numpy as np

from misc_postprocessing import *

# rc('font', **{'family': 'serif', 'serif': ['Computer Modern']})
# rc('text', usetex=True)
```

```python
discretization_list = [8, 16, 32, 64, 128, 256, 512, 1024]
# List of 'neval' values
neval_values = [(i, i) for i in discretization_list]
latent_dim = 9
```

```python
# Initialize an empty list to store the results
runtime_per_iter_data = []

# Loop over each 'neval_t' and 'neval_x' value
for neval_t, neval_x in neval_values:
    # Paths for all notebooks, moving one folder backward
    a_PI_Vanilla_NO_Reverse_AD_tiling_path = f'../results/ReverseAD_vs_ForwardAD/a_PI-Vanilla-NO_Reverse_AD-tiling/output_neval_t={neval_t}_neval_x={neval_x}.ipynb'
    b_PI_Latent_NO_with_PCA_Reverse_AD_tiling_path = f'../results/ReverseAD_vs_ForwardAD/b_PI-Latent-NO_with-PCA_Reverse_AD-tiling/output_neval_t={neval_t}_neval_x={neval_x}_latent_dim={latent_dim}.ipynb'
    a_PI_Vanilla_NO_Forward_AD_einsum_path = f'../results/ReverseAD_vs_ForwardAD/a_PI-Vanilla-NO_Forward_AD-einsum/output_neval_t={neval_t}_neval_x={neval_x}.ipynb'
    b_PI_Latent_NO_with_PCA_Forward_AD_einsum_path = f'../results/ReverseAD_vs_ForwardAD/b_PI-Latent-NO_with-PCA_Forward_AD-einsum/output_neval_t={neval_t}_neval_x={neval_x}_latent_dim={latent_dim}.ipynb'

    # Extract runtime and memory for all notebooks
    a_PI_Vanilla_NO_Reverse_AD_tiling_runtime, a_PI_Vanilla_NO_Reverse_AD_tiling_memory = extract_runtime_and_memory_from_notebook(a_PI_Vanilla_NO_Reverse_AD_tiling_path)
    b_PI_Latent_NO_with_PCA_Reverse_AD_tiling_runtime, b_PI_Latent_NO_with_PCA_Reverse_AD_tiling_memory = extract_runtime_and_memory_from_notebook(b_PI_Latent_NO_with_PCA_Reverse_AD_tiling_path)
    a_PI_Vanilla_NO_Forward_AD_einsum_runtime, a_PI_Vanilla_NO_Forward_AD_einsum_memory = extract_runtime_and_memory_from_notebook(a_PI_Vanilla_NO_Forward_AD_einsum_path)
    b_PI_Latent_NO_with_PCA_Forward_AD_einsum_runtime, b_PI_Latent_NO_with_PCA_Forward_AD_einsum_memory = extract_runtime_and_memory_from_notebook(b_PI_Latent_NO_with_PCA_Forward_AD_einsum_path)

    # Append the results to the list, with ReverseAD-tiling first, then ForwardAD-einsum
    runtime_per_iter_data.append({
        'neval_t': neval_t,
        'neval_x': neval_x,
        'PI-Vanilla-NO (ReverseAD-tiling) Runtime (sec/iter)': a_PI_Vanilla_NO_Reverse_AD_tiling_runtime,
        'PI-Vanilla-NO (ReverseAD-tiling) Memory (MB)': a_PI_Vanilla_NO_Reverse_AD_tiling_memory,
        'PI-Latent-NO (with-PCA) (ReverseAD-tiling) Runtime (sec/iter)': b_PI_Latent_NO_with_PCA_Reverse_AD_tiling_runtime,
        'PI-Latent-NO (with-PCA) (ReverseAD-tiling) Memory (MB)': b_PI_Latent_NO_with_PCA_Reverse_AD_tiling_memory,
        'PI-Vanilla-NO (ForwardAD-einsum) Runtime (sec/iter)': a_PI_Vanilla_NO_Forward_AD_einsum_runtime,
        'PI-Vanilla-NO (ForwardAD-einsum) Memory (MB)': a_PI_Vanilla_NO_Forward_AD_einsum_memory,
        'PI-Latent-NO (with-PCA) (ForwardAD-einsum) Runtime (sec/iter)': b_PI_Latent_NO_with_PCA_Forward_AD_einsum_runtime,
        'PI-Latent-NO (with-PCA) (ForwardAD-einsum) Memory (MB)': b_PI_Latent_NO_with_PCA_Forward_AD_einsum_memory
    })

# Store the runtime and memory data in a DataFrame
df = pd.DataFrame(runtime_per_iter_data)

# Set the index name
df.index.name = 'Evaluation'

# Display the DataFrame
df.style.set_table_styles([dict(selector="th", props=[('max-width', '100px')])])

# Display the DataFrame
df
```

```python
display(df)

### RUNTIME (sec/iter)

# Assuming df is the DataFrame we created
n_values = df['neval_t']

# Plotting the four runtime columns
plt.figure(figsize=(10, 6))

plt.plot(n_values, df['PI-Vanilla-NO (ReverseAD-tiling) Runtime (sec/iter)'], label='PI-Vanilla-NO (ReverseAD-tiling)', 
         marker='o', linestyle='--', color='blue')
plt.plot(n_values, df['PI-Latent-NO (with-PCA) (ReverseAD-tiling) Runtime (sec/iter)'], label=f'PI-Latent-NO (with-PCA, latent-dim = {latent_dim}) (ReverseAD-tiling)', 
         marker='o', linestyle='--', color='red')
plt.plot(n_values, df['PI-Vanilla-NO (ForwardAD-einsum) Runtime (sec/iter)'], label='PI-Vanilla-NO (ForwardAD-einsum)', 
         marker='o', linestyle='-', color='blue')
plt.plot(n_values, df['PI-Latent-NO (with-PCA) (ForwardAD-einsum) Runtime (sec/iter)'], label=f'PI-Latent-NO (with-PCA, latent-dim ={latent_dim}) (ForwardAD-einsum)', 
         marker='o', linestyle='-', color='red')

# Setting plot details
plt.xscale('log')  # Logarithmic scale for x-axis
plt.yscale('log') # Logarithmic scale for y-axis
plt.xlabel('Discretization of the solution space ($n_c^2$)') # Number of collocation points ($n_c^2$)
plt.ylabel('Runtime (sec/iter)')
plt.xticks(n_values, [f'{int(n)}$^2$' for n in n_values])  # Custom x-axis labels with superscript
plt.legend(loc='best')
plt.grid(True)
plt.show()


### Memory (MB)

# Assuming df is the DataFrame we created
n_values = df['neval_t']

# Plotting the four runtime columns
plt.figure(figsize=(10, 6))

plt.plot(n_values, df['PI-Vanilla-NO (ReverseAD-tiling) Memory (MB)'], label='PI-Vanilla-NO (ReverseAD-tiling)', 
         marker='o', linestyle='--', color='blue')
plt.plot(n_values, df['PI-Latent-NO (with-PCA) (ReverseAD-tiling) Memory (MB)'], label=f'PI-Latent-NO (with-PCA, latent-dim = {latent_dim}) (ReverseAD-tiling)', 
         marker='o', linestyle='--', color='red')
plt.plot(n_values, df['PI-Vanilla-NO (ForwardAD-einsum) Memory (MB)'], label='PI-Vanilla-NO (ForwardAD-einsum)', 
         marker='o', linestyle='-', color='blue')
plt.plot(n_values, df['PI-Latent-NO (with-PCA) (ForwardAD-einsum) Memory (MB)'], label=f'PI-Latent-NO (with-PCA, latent-dim = {latent_dim}) (ForwardAD-einsum)', 
         marker='o', linestyle='-', color='red')

# Setting plot details
plt.xscale('log')  # Logarithmic scale for x-axis
plt.yscale('log') # Logarithmic scale for y-axis
plt.xlabel('Discretization of the solution space ($n_c^2$)') # Number of collocation points ($n_c^2$)
plt.ylabel('Memory (MB)')
plt.xticks(n_values, [f'{int(n)}$^2$' for n in n_values])  # Custom x-axis labels with superscript
plt.legend(loc='best')
plt.grid(True)
plt.show()
```
