```python
import nbformat
import pandas as pd
import numpy as np
from matplotlib import rc
import matplotlib.pyplot as plt
import numpy as np

from misc_postprocessing import *

# rc('font', **{'family': 'serif', 'serif': ['Computer Modern']})
# rc('text', usetex=True)
```

```python
# List of 'neval' values
neval_x_values = [8, 16, 32, 64, 128, 256, 512, 1024, 2048, 4096]
neval_values = [(8, i) for i in neval_x_values] + [(64, i) for i in neval_x_values]
latent_dim = 9
```

```python
# Initialize an empty list to store the results
runtime_per_iter_data = []

# Loop over each 'neval_t' and 'neval_x' value
for neval_t, neval_x in neval_values:
    # Paths for all notebooks, moving one folder backward
    a_PI_Vanilla_NO_Forward_AD_einsum_path = f'../results/ForwardAD_neval_t_constant/a_PI-Vanilla-NO_Forward_AD-einsum/output_neval_t={neval_t}_neval_x={neval_x}.ipynb'
    b_PI_Latent_NO_with_PCA_Forward_AD_einsum_path = f'../results/ForwardAD_neval_t_constant/b_PI-Latent-NO_with-PCA_Forward_AD-einsum/output_neval_t={neval_t}_neval_x={neval_x}_latent_dim={latent_dim}.ipynb'

    # Extract runtime and memory for all notebooks
    a_PI_Vanilla_NO_Forward_AD_einsum_runtime, a_PI_Vanilla_NO_Forward_AD_einsum_memory = extract_runtime_and_memory_from_notebook(a_PI_Vanilla_NO_Forward_AD_einsum_path)
    b_PI_Latent_NO_with_PCA_Forward_AD_einsum_runtime, b_PI_Latent_NO_with_PCA_Forward_AD_einsum_memory = extract_runtime_and_memory_from_notebook(b_PI_Latent_NO_with_PCA_Forward_AD_einsum_path)

    # Append the results to the list, with ReverseAD-tiling first, then ForwardAD-einsum
    runtime_per_iter_data.append({
        'neval_t': neval_t,
        'neval_x': neval_x,
        'PI-Vanilla-NO (ForwardAD-einsum) Runtime (sec/iter)': a_PI_Vanilla_NO_Forward_AD_einsum_runtime,
        'PI-Vanilla-NO (ForwardAD-einsum) Memory (MB)': a_PI_Vanilla_NO_Forward_AD_einsum_memory,
        'PI-Latent-NO (with-PCA) (ForwardAD-einsum) Runtime (sec/iter)': b_PI_Latent_NO_with_PCA_Forward_AD_einsum_runtime,
        'PI-Latent-NO (with-PCA) (ForwardAD-einsum) Memory (MB)': b_PI_Latent_NO_with_PCA_Forward_AD_einsum_memory
    })

# Store the runtime and memory data in a DataFrame
df_all = pd.DataFrame(runtime_per_iter_data)

# Set the index name
df_all.index.name = 'Evaluation'

# Display the DataFrame
df_all.style.set_table_styles([dict(selector="th", props=[('max-width', '100px')])])

# Display the DataFrame
df_all
```

```python
df = df_all.groupby('neval_t').get_group(8)
display(df)

### RUNTIME (sec/iter)

# Plotting the four runtime columns
plt.figure(figsize=(10, 6))

plt.plot(df['neval_x'], df['PI-Vanilla-NO (ForwardAD-einsum) Runtime (sec/iter)'], label='PI-Vanilla-NO', 
         marker='o', linestyle='-', color='blue')
plt.plot(df['neval_x'], df['PI-Latent-NO (with-PCA) (ForwardAD-einsum) Runtime (sec/iter)'], label=f'PI-Latent-NO (with-PCA, latent-dim ={latent_dim})', 
         marker='o', linestyle='-', color='red')

# Setting plot details
plt.xscale('log')  # Logarithmic scale for x-axis
# plt.yscale('log') # Logarithmic scale for y-axis
plt.xlabel('Discretization along x axis ($neval_x$)') 
plt.ylabel('Runtime (sec/iter)')
plt.xticks(df['neval_x'], [f'{int(n)}' for n in df['neval_x']])  # Custom x-axis labels with superscript
plt.legend(loc='best')
plt.grid(True)
plt.title(f'ForwardAD-einsum, $neval_t={8}$')
plt.show()


### Memory (MB)

# Plotting the four runtime columns
plt.figure(figsize=(10, 6))

plt.plot(df['neval_x'], df['PI-Vanilla-NO (ForwardAD-einsum) Memory (MB)'], label='PI-Vanilla-NO', 
         marker='o', linestyle='-', color='blue')
plt.plot(df['neval_x'], df['PI-Latent-NO (with-PCA) (ForwardAD-einsum) Memory (MB)'], label=f'PI-Latent-NO (with-PCA, latent-dim = {latent_dim})', 
         marker='o', linestyle='-', color='red')

# Setting plot details
plt.xscale('log')  # Logarithmic scale for x-axis
# plt.yscale('log') # Logarithmic scale for y-axis
plt.xlabel('Discretization along x axis ($neval_x$)') 
plt.ylabel('Memory (MB)')
plt.xticks(df['neval_x'], [f'{int(n)}' for n in df['neval_x']])  # Custom x-axis labels with superscript
plt.legend(loc='best')
plt.grid(True)
plt.title(f'ForwardAD-einsum, $neval_t={8}$')
plt.show()
```
