```python
import time
import torch
import functorch
import warnings
warnings.filterwarnings('ignore')

# Define a sample function
def f(x):
    return x**3 + (2 * x**2) + (87*x**4)

# Inputs
primals = torch.tensor([2.0], requires_grad=True)
tangents = torch.tensor([1.0])

# Combined first and second derivatives
def FWDAD_first_and_second_order_derivative(f, primals, tangents):
    g = lambda primals: functorch.jvp(f, (primals,), (tangents,))[1]  # Computes first derivative of f
    a, tangents_out = functorch.jvp(g, (primals,), (tangents,))  # Computes second derivative of f
    return a, tangents_out

# Separate functions
def FWDAD_first_order_derivative(f, primals, tangents):
    _, tangents_out = functorch.jvp(f, (primals,), (tangents,))
    return tangents_out

def FWDAD_second_order_derivative(f, primals, tangents):
    g = lambda primals: functorch.jvp(f, (primals,), (tangents,))[1]
    _, tangents_out = functorch.jvp(g, (primals,), (tangents,))
    return tangents_out


# Measure time for the combined function
start_time = time.time()
first, second = FWDAD_first_and_second_order_derivative(f, primals, tangents)
combined_time = time.time() - start_time
print(first, second)

# Measure time for the separate functions
start_time = time.time()
first_order_result = FWDAD_first_order_derivative(f, primals, tangents)
second_order_result = FWDAD_second_order_derivative(f, primals, tangents)
separate_time = time.time() - start_time
print(first_order_result, second_order_result)

# Output results
print(f"Combined function time: {combined_time:.10f} seconds")
print(f"Separate functions time: {separate_time:.10f} seconds")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
# import config as cfg
from scipy.optimize import newton
import matplotlib
from mpl_toolkits.axes_grid1 import make_axes_locatable
from matplotlib import ticker

matplotlib.rcParams['text.usetex'] = True
plt.rcParams.update({'font.size': 35})

x = np.linspace(-np.pi, np.pi, num=100, dtype=np.float32)
y = np.linspace(-np.pi, np.pi, num=100, dtype=np.float32)
X, Y = np.meshgrid(x, y, indexing='ij')
U = np.sin(X) * np.sin(Y)

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(6.5, 5), sharey=True, layout='constrained')
c0 = ax.pcolormesh(X, Y, U, vmin=-1, vmax=1, edgecolors='face', rasterized=True, cmap='coolwarm')
ax.set_xlabel(r'$x$')
ax.set_ylabel(r'$y$')
ax.set_xticks([-np.pi, 0, np.pi], [r'$-\pi$', 0, r'$\pi$'])
ax.set_yticks([-np.pi, 0, np.pi], [r'$-\pi$', 0, r'$\pi$'])
cbar1 = fig.colorbar(c0, ax=ax, shrink=0.75)
cbar1.formatter.set_powerlimits((0, 0))
cbar1.ax.yaxis.set_offset_position('left')
tick_locator = ticker.MaxNLocator(nbins=5)
cbar1.locator = tick_locator
cbar1.update_ticks()
plt.show()
```
